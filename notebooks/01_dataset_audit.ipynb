```python
haxby = datasets.fetch_haxby()
```

```python
print("Functional scans:", len(haxby.func))
print("Anatomical scans:", len(haxby.anat))
```

```python
print("First functional file:")
print(haxby.func[0])

print("\nFirst anatomical file:")
print(haxby.anat[0])
```

```python
print("Total timepoints:", len(labels))
print(labels["labels"].value_counts())
```
